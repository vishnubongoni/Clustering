```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
#Import Dataset
data=pd.read_excel("EastWestAirlines.xlsx",sheet_name='data')
data
```

```python
data2=data.drop(['ID#'],axis=1)
data2
```

```python
data2.info()
```

```python
#Normalize the heterogenous numerical by standardscaler
data2_norm=StandardScaler().fit_transform(data2)
```

```python
# Use Elbow Graph to find optimum number of  clusters (K value) from K values range
# The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion WCSS
# random state can be anything from 0 to 42, but the same number to be used everytime,so that the results don't change.
```

```python
# within-cluster sum-of-squares criterion
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(data2_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## Build Cluster Algorithm using K=4

```python
# Cluster algorithm using K=4
clusters4=KMeans(4,random_state=30).fit(data2_norm)
clusters4
```

```python
clusters4.labels_
```

```python
# Assign clusters to the data set
data4=data2.copy()
data4["clusters4id"]=clusters4.labels_
data4
```

```python
clusters4.cluster_centers_
```

```python
# Group data by Clusters (K=4)
data4.groupby("clusters4id").agg(["mean"]).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))
plt.scatter(data4['clusters4id'],data4['Balance'], c=clusters4.labels_)
```
