HIERARCHICAL

```python
# Normalization function
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
crime=(data.iloc[:,1:])
data
```

```python
# Normalized data frame (considering the numerical part of data)
crime_norm = norm_func(crime.iloc[:,1:])
```

```python
crime_norm
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(crime_norm, method='single'))
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(crime_norm, method='average'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'single')
```

```python
clusters=pd.DataFrame(hc.fit_predict(crime_norm),columns=['Clusters'])
clusters['Clusters'].value_counts()
```

```python
# save clusters for chart
y_hc = hc.fit_predict(crime_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
Clusters
```

```python
crime['Clusters']=clusters
```

```python
# Normalization function
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)

# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(crime.iloc[:,1:])


df_norm.head(10)  # Top 10 rows
```

```python
# Selecting 5 clusters from the above scree plot which is the optimum number of clusters
model=KMeans(n_clusters=5)
model.fit(df_norm)

model.labels_ # getting the labels of clusters assigned to each row
```

```python
md=pd.Series(model.labels_)  # converting numpy array into pandas series object
crime['clust']=md # creating a  new column and assigning it to new column
df_norm.head()
```

```python
crime.iloc[:,1:7].groupby(crime.clust).mean()
```

```python
crime.head()
```

```python
Wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(df_norm)
    Wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),Wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('Wcss')
plt.show()
```

DBSCAN

```python
crime=pd.read_csv("crime_data.csv")
crime.head()
```

```python
crime.info()
```

```python
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
array=crime.values
```

```python
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
```

```python
X
```

```python
dbscan = DBSCAN(eps=0.9, min_samples=5)
dbscan.fit(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```

```python
pd.concat([crime,cl],axis=1)
```
